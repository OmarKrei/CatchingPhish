```python
import streamlit as st
import joblib
import pandas as pd

# Load model
model = joblib.load("xgb_model.pkl")

# UI
st.title("Phishing URL Classifier")
st.write("Enter a URL below to check if it's legitimate or phishing.")

# Input field
url_input = st.text_input("Enter URL")

# Extract features from the input URL (stub example — you must replace with your real preprocessing logic)
def extract_features(url):
    # Replace this with your actual feature extraction steps
    features = {
        'Length': len(url),
        'HasHTTPS': int('https' in url),
        'DotCount': url.count('.'),
        # Add all other features expected by the model...
    }
    return pd.DataFrame([features])

# Predict
if st.button("Classify"):
    features_df = extract_features(url_input)
    prediction = model.predict(features_df)[0]
    
    result = "Legitimate ✅" if prediction == 1 else "Phishing ⚠️"
    st.subheader(f"Prediction: {result}")
```
