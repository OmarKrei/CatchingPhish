## Model 1 - Random Forest - Iteration 1

````python
# splitting independent & dependent variables
X = phish.drop(columns=['label'])
y = old_phish['label']

# splitting train & test sets 
X_train, X_test, y_train, y_test = train_test_split( 
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42)

# building a pipeline with random forest
rf_pipeline = Pipeline([
    ('clf', RandomForestClassifier(random_state=42))
])

# Optimizing the parameters in iteration 1
rf_param_grid_iter1 = {
    'clf__n_estimators': [250, 300, 350],         # around best
    'clf__max_depth': [18, 20, 22],               # close to 20
    'clf__min_samples_split': [2, 3, 4],          # fine-tuning split
    'clf__min_samples_leaf': [1, 2, 3]            # close to 2
}

# running a grid search
rf_grid = GridSearchCV(
    rf_pipeline,
    rf_param_grid_iter1,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# fitting the model
rf_grid.fit(X_train, y_train)

# evaluating the model
y_pred_rf = rf_grid.predict(X_test)
y_proba_rf = rf_grid.predict_proba(X_test)[:, 1]
````

````python
print("Best Parameters (baseline):\n", rf_grid.best_params_)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba_rf))
````

### Classification Report Analysis - Iteration 1

Progress has been made, the updated parameters have improved percision and accuracy by decreasing both the number of false positives and negatives. To reiterate, the original baseline has an incredible performance, but, it's good to see that in the iteration we were able to improve the confusion matrix distribution even if it's by a tiny amount. 

<br> We can see that the best parameters have changed across the 2 models. the max depth has increased to 22 from 20, minimum samples/leaf has decreased to 1, the minimum split has increased to 3 from 2, and the estimators have decreased to 250 from 300. 

<br> Given that every single parameter has changed, it makes sense to run another iteration to see if these are the best params. Although, the model recall, precision and F1-socre will stay the same, it's important to see that we've gotten the best model possible. 

## Model 1 - Random Forest - Iteration 2 (Final)

````python
# splitting independent & dependent variables
X = phish.drop(columns=['label'])
y = old_phish['label']

# splitting train & test sets 
X_train, X_test, y_train, y_test = train_test_split( 
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42)

# building a pipeline with random forest
rf_pipeline = Pipeline([
    ('clf', RandomForestClassifier(random_state=42))
])

rf_param_grid_final = {
    'clf__n_estimators': [250, 300],         # around best
    'clf__max_depth': [20, 22, 24],               # Updated max depth given best param was 22
    'clf__min_samples_split': [2, 3, 4],          # fine-tuning split
    'clf__min_samples_leaf': [1, 2]            # close to 2
}

# running a grid search
rf_grid = GridSearchCV(
    rf_pipeline,
    rf_param_grid_final,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# fitting the model
rf_grid.fit(X_train, y_train)

# evaluating the model
y_pred_rf = rf_grid.predict(X_test)
y_proba_rf = rf_grid.predict_proba(X_test)[:, 1]
````

````python
print("Best Parameters (baseline):\n", rf_grid.best_params_)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba_rf))
````

### Classification Report Analysis - Iteration 2

As we can see, our model has been optimized. The best parameters have remained the same across the board, when comparing to iteration 1.

<br> This will be the benchmark model that we compare with the XGBoost model. The better performing model will be moved to production.

## Model 2 - XGBoost - Iteration 1

````python
# splitting independent & dependent variables

X = phish.drop(columns=['label'])
y = old_phish['label']

# Split (same stratified setup)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

xgb_pipeline = Pipeline([
    ('clf', xgb.XGBClassifier(
        objective='binary:logistic',
        eval_metric='auc',
        use_label_encoder=False,
        random_state=42
    ))
])

xgb_param_grid_iter1 = {
    'clf__n_estimators': [150, 200, 250],    # Number of trees
    'clf__max_depth': [5, 6, 7],             # Tree complexity
    'clf__learning_rate': [0.1, 0.15, 0.2],  # smaller = slower learning, better generalization
    'clf__subsample': [0.9, 1.0],            # Row subsampling for regularization
    'clf__colsample_bytree': [0.6, 0.7, 0.8] # Column subsampling (helps avoid overfitting)
}
# Grid search
xgb_grid = GridSearchCV(
    xgb_pipeline,
    param_grid=xgb_param_grid_iter1,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

xgb_grid.fit(X_train, y_train)
````

### XGBoost Model Result - No L1 or L2 Regularization

````python
y_pred_xgb = xgb_grid.predict(X_test)
y_proba_xgb = xgb_grid.predict_proba(X_test)[:, 1]

print('Best Parameters:', xgb_grid.best_params_)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_xgb))
print('Classification Report:\n', classification_report(y_test, y_pred_xgb))
print('ROC AUC Score:', roc_auc_score(y_test, y_proba_xgb))
````

There is a slight improvement in the performance of the model when the best params were altered to the following parameters: 

<br>clf__max+depth 6 --> 5
- This change made the tree shallower reducing complexity and overfitting. 
- This made the model more generalized, avoiding overly spcific patterns that didn't translate well to unseen data.

-----
<br>clf__n_estimatros 200 --> 250 
- The counter balance was increasing the number of estimatros. It added more trees to allow the model to refine predicitons across more iterations. 
- The difference is the model made smaller, incremental improvements, this was even more pronounced with the learning rate set at 0.2. 

# L1 & L2 Regularization

Regularization is a technique to prevent overfitting by penalizing large of complex model weights. 

----
#### L1 Regularization
Also called Lasso Regression, adds the absolute values of the weights to the loss function. 

<br> The effect of this is that the model is encourages to shrink weights to excatly zero, helping the model select features more optimally and more simple. This can be helpful if you want a quick model and are not worried about dropping features. 

In the case of this project, making the model less percise can be seen as a problem, however it would not hurt to see its performance. 

----
#### L2 Regularization 

Also called Ridge Regression, adds the squared values fo the weights to the loss. The impact of this, is that the weights of features is driven to be smaller, spreading the influence across all features. The benefit of this is that it helps prevent extreme values, reducing overfitting but still keeping all the features in the model. 



## Model 2 - XGBoost - Iteration 1 - W/ L1 Regularization

````python
# splitting independent & dependent variables
X = phish.drop(columns=['label'])
y = old_phish['label']

# building a pipeline with random forest
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# building a pipeline with l1 regularization
xgb_pipeline_l1 = Pipeline([
    ('clf', XGBClassifier(
        objective='binary:logistic',
        eval_metric='auc',
        use_label_encoder=False,
        reg_alpha=1.0,              # adding L1 Regularization strength
        reg_lambda=0.0,             # confirming no L2 regularization
        random_state=42
    ))
])

xgb_param_grid_iter2 = {
    'clf__n_estimators': [200, 250],    # Number of trees
    'clf__max_depth': [6, 7],             # Tree complexity
    'clf__learning_rate': [0.15, 0.2],  # smaller = slower learning, better generalization
    'clf__subsample': [0.9, 1.0],            # Row subsampling for regularization
    'clf__colsample_bytree': [0.6, 0.7, 0.8] # Column subsampling (helps avoid overfitting)
}
# running a grid search
xgb_grid_l1 = GridSearchCV(
    xgb_pipeline_l1,
    param_grid=xgb_param_grid_iter2,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# fitting the model
xgb_grid_l1.fit(X_train, y_train)
````

### XGBoost Model Result - W/L1 Regularization

````python
y_pred_xgb = xgb_grid_l1.predict(X_test)
y_proba_xgb = xgb_grid_l1.predict_proba(X_test)[:, 1]

print('Best Parameters:', xgb_grid_l1.best_params_)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_xgb))
print('Classification Report:\n', classification_report(y_test, y_pred_xgb))
print('ROC AUC Score:', roc_auc_score(y_test, y_proba_xgb))
````

<br> The L1- regularized model is slightly more conservative, using fewer trees (250 --> 200) and a lower learning rate (0.2 --> 0.15). 

<br> Furthermore, with L1 regularization the model goes slightly deeper looking/allowing more complex splits. 

<br> The critical comparison in performacne is that the L1 containing model has a higher AUC, but 36% more false negatives. When the margins are so tight, given the predicting power of the features, this increase in false negatives is unaccepatable. 

## Model 2 - XGBoost - Iteration 1 - W/ L2 Regularization

````python
# splitting independent & dependent variables
X = phish.drop(columns=['label'])
y = old_phish['label']

# building a pipeline with random forest
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# building a pipeline with L2 regularization
xgb_pipeline_l2 = Pipeline([
    ('clf', XGBClassifier(
        objective='binary:logistic',
        eval_metric='auc',
        use_label_encoder=False,
        reg_alpha=0.0,              # adding L1 Regularization strength
        reg_lambda=1.0,             # confirming no L2 regularization
        random_state=42
    ))
])

xgb_param_grid_iter3 = {
    'clf__n_estimators': [200, 250],    # Number of trees
    'clf__max_depth': [6, 7],             # Tree complexity
    'clf__learning_rate': [0.15, 0.2],  # smaller = slower learning, better generalization
    'clf__subsample': [0.9, 1.0],            # Row subsampling for regularization
    'clf__colsample_bytree': [0.6, 0.7, 0.8] # Column subsampling (helps avoid overfitting)
}
# running a grid search
xgb_grid_l2 = GridSearchCV(
    xgb_pipeline_l2,
    param_grid=xgb_param_grid_iter3,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# fitting the model
xgb_grid_l2.fit(X_train, y_train)

````

### XGBoost Model Result - W/L2 Regularization

````python
y_pred_xgb = xgb_grid_l2.predict(X_test)
y_proba_xgb = xgb_grid_l2.predict_proba(X_test)[:, 1]

print('Best Parameters:', xgb_grid_l2.best_params_)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_xgb))
print('Classification Report:\n', classification_report(y_test, y_pred_xgb))
print('ROC AUC Score:', roc_auc_score(y_test, y_proba_xgb))
````

### Shapley Additive Explanations (SHAP)

SHAP summary plot and it shows how the XGBoost model is making decisions. Each dot in the SHAP plot is one predicition. the X-Axis (the SHAP value) is how much the feature push the model towards predicting phishing. The Y-axis are the features. The colour shows how much of that feature there is. So blue means a lower quantity, whereas red means higher. 

````python
# creating SHAP explainer for XGBoost model
explainer = shap.Explainer(xgb_model)

# using x_test for SHAP values to show how the model classified for the test dataset
shap_values = explainer(X_test)

# plotting summary
shap.plots.beeswarm(shap_values)
````

### SHAP Analysis

As discussed before and as we can see from the SHAP plot, LineOfCode, NoOfExternalRef, and NoOfImage are amongst the highest predicters of phishing vs legitimate URL's/Websites. 

<br> LineOfCode: We can see that the URL's with the lowest feature value are spread across the X-axis, however, the URL's with the highest feature value are all concentrated towards the positive right side of the SHAP plot. This shows that our previous conclusion is correct. The more LinesOfCode, the more likely the URL is legitimate (hence why the red dots are concentrated on the right hand side of the plot). Whereas the phishing URL's are spread across, and concentrated towards the left hand side, meaning a lower quantity of lines of code. 

<br> NoOfExternalRef & NoOfImages: The same can be said about the NoofExternalRef and NoOfImages, the feweres the external referecnes and images the more likely the URL is phishing. To reiterate, this is because scammers are not willing to put in the effort and time to populate their websites. It's meant to be used and then dumped, whereas real websites are owned by legitimate businesses that are looking to use their website for as long as they're open.
